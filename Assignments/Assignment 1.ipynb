### Q.1
> #### Notes for output
>
> Test value considered: (x,y: x=y; x={1,2,3,4,5})

```python
from matplotlib import pyplot as plt

def get_data_points():
    x_values = []
    y_values = []

    # Loop to get at least 5 data points
    for i in range(5):
        while True:
            try:
                # User input for x and y coordinates
                x = float(input(f"Enter x-coordinate for point {i+1}: "))
                y = float(input(f"Enter y-coordinate for point {i+1}: "))
                x_values.append(x)
                y_values.append(y)
                break
            except ValueError:
                print("Invalid input. Please enter numeric values for x and y coordinates.")
    return x_values, y_values

# Get data points from the user
x_data, y_data = get_data_points()

# Plot the line graph
plt.plot(x_data, y_data, marker='o', linestyle='-', color='b')

# Adding title and labels
plt.title("Line Graph of Given Data Points")
plt.xlabel("X")
plt.ylabel("Y")

# Displaying grid
plt.grid(True)

# Display the graph
plt.show()
```

```python
def traffic_signal():
    # Get the color input from the user
    signal_color = input("Enter the traffic signal color (Red, Yellow, or Green): ").strip().lower()

    # Conditional statements to determine the action
    if signal_color == "red":
        print("Action: Stop!")
    elif signal_color == "yellow":
        print("Action: Prepare to stop or go!")
    elif signal_color == "green":
        print("Action: Go!")
    else:
        print("Invalid input! Please enter Red, Yellow, or Green.")

# Call the function
traffic_signal()
```

### Q.3
> #### Notes for output
>
> Test value considered: 1,20,2,19,5,10,8,8,7,6 (* 6 was the correct answer; attempts = 10)

```python
import random

def guess_the_number():
    # Generate a random number between 1 and 20
    secret_number = random.randint(1, 20)

    # Initialize the number of attempts
    attempts = 0
    guessed_correctly = False

    print("Welcome to the Guessing Game!")
    print("I have selected a number between 1 and 20. Try to guess it!")

    # Loop until the user guesses correctly
    while not guessed_correctly:
        try:
            # Get the user's guess
            guess = int(input("Enter your guess: "))
            attempts += 1

            # Check if the guess is correct
            if guess < secret_number:
                print("Too low! Try again.")
            elif guess > secret_number:
                print("Too high! Try again.")
            else:
                print(f"Correct! The number was {secret_number}.")
                print(f"You guessed it in {attempts} attempts!")
                guessed_correctly = True
        except ValueError:
            print("Please enter a valid number.")

# Call the function to start the game
guess_the_number()
```

### Q.4
> #### Notes for output
>
> Test value considered: example

```python
def count_vowels():
    # Get the string input from the user
    user_input = input("Enter a string: ")

    # Define the set of vowels (both uppercase and lowercase)
    vowels = "aeiouAEIOU"

    # Initialize the count of vowels
    vowel_count = 0

    # Loop through each character in the string
    for char in user_input:
        if char in vowels:
            vowel_count += 1

    # Output the total number of vowels
    print(f"The total number of vowels in the string is: {vowel_count}")

# Call the function
count_vowels()
```

### Q.5
> #### Notes for output
>
> Test value considered: Vegetables (Mixed)*100

```python
# Grocery items and their prices
grocery_items = {
    "Rice (5kg)": 300,
    "Wheat Flour (5kg)": 250,
    "Sugar (2kg)": 120,
    "Milk (1L)": 60,
    "Eggs (12 pcs)": 90,
    "Bread (1 pack)": 40,
    "Butter (500g)": 150,
    "Cheese (500g)": 180,
    "Tea (250g)": 110,
    "Coffee (200g)": 200,
    "Cooking Oil (1L)": 250,
    "Vegetables (Mixed)": 180
}

# Function to calculate the total bill
def grocery_billing():
    print("Welcome to the Grocery Store!")
    total_bill = 0
    selected_items = []

    # Display available items and their prices
    print("\nAvailable items and prices:")
    for item, price in grocery_items.items():
        print(f"{item}: ₹{price}")

    while True:
        # Get the user's item choice
        item_choice = input("\nEnter the item you want to buy (or type 'done' to finish): ").strip()

        if item_choice.lower() == 'done':
            break

        # Check if the item is in the available items list
        if item_choice in grocery_items:
            quantity = int(input(f"Enter the quantity for {item_choice}: "))
            total_bill += grocery_items[item_choice] * quantity
            selected_items.append(f"{item_choice} (x{quantity})")
        else:
            print("Invalid item. Please enter a valid item name from the list.")

    # Display the total bill before discount
    print("\nYour selected items:")
    for item in selected_items:
        print(item)

    print(f"\nTotal Bill: ₹{total_bill}")

    # Apply discount if the total bill is above ₹500
    if total_bill > 500:
        discount = total_bill * 0.10
        discounted_price = total_bill - discount
        print(f"Discount Applied: ₹{discount:.2f}")
        print(f"Total after Discount: ₹{discounted_price:.2f}")
    else:
        print("No discount applied.")

# Call the function to run the program
grocery_billing()
```
