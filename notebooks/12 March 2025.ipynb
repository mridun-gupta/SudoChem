# Question
### Code a simple calculator to perform the following six major operations:
- Addition ;
- Subtraction ;
- Multiplication ;
- Division ;
- Squaring a number ;
- Square-rooting a number ;
- Inter-conversion of degree and radians.

```python
import math

# Addition Function
def add(a, b):
    return a + b

# Subtraction Function
def subtract(a, b):
    return a - b

# Multiplication Function
def multiply(a, b):
    return a * b

# Division Function
def divide(a, b):
    q = a / b
    r = a % b
    if b == 0:
        return "Error! Division by zero."
    else:
        return f"\nQuotient is: {q}\nRemainder is: {r}"

# Function to square a number
def square(a):
    return a ** 2

# Function to square-root a number
def square_root(a):
    if a < 0:
        return "Error! Cannot compute square root of a negative number."
    return math.sqrt(a)

# Functions to inter-conversion of degrees and radians
def degrees_to_radians(deg):
    return math.radians(deg)
def radians_to_degrees(rad):
    return math.degrees(rad)

# Main Calculator Function
def calculator():
    while True:
        print("1. Addition")
        print("2. Subtraction")
        print("3. Multiplication")
        print("4. Division")
        print("5. Square a number")
        print("6. Square root of a number")
        print("7. Convert Degrees to Radians")
        print("8. Convert Radians to Degrees")
        print("9. Exit")

        choice = input("Enter your choice (1-9): ")
        if choice == '1':
            a = float(input("Enter first number: "))
            b = float(input("Enter second number: "))
            print("Result:", add(a, b))
        elif choice == '2':
            a = float(input("Enter first number: "))
            b = float(input("Enter second number: "))
            print("Result:", subtract(a, b))
        elif choice == '3':
            a = float(input("Enter first number: "))
            b = float(input("Enter second number: "))
            print("Result:", multiply(a, b))
        elif choice == '4':
            a = float(input("Enter numerator: "))
            b = float(input("Enter denominator: "))
            print("Result:", divide(a, b))
        elif choice == '5':
            a = float(input("Enter the number: "))
            print("Result:", square(a))
        elif choice == '6':
            a = float(input("Enter the number: "))
            print("Result:", square_root(a))
        elif choice == '7':
            deg = float(input("Enter degrees: "))
            print("Result:", degrees_to_radians(deg), "radians")
        elif choice == '8':
            rad = float(input("Enter radians: "))
            print("Result:", radians_to_degrees(rad), "degrees")
        elif choice == '9':
            print("Exiting calculator. Goodbye!")
            break
        else:
            print("Invalid input! Please enter a number between 1 and 9.")

# Run the calculator
calculator()
```

### Let's optimize this version.

```python
import math

# Basic Arithmetic Functions
def add(a, b): return a + b
def subtract(a, b): return a - b
def multiply(a, b): return a * b
def divide(a, b):
    if b == 0:
        return "Error! Division by zero."
    return f"Quotient: {a / b}, Remainder: {a % b}"

# Advanced Functions
def square(a): return a ** 2
def square_root(a): return math.sqrt(a) if a >= 0 else "Error! Cannot compute square root of a negative number."
def degrees_to_radians(deg): return math.radians(deg)
def radians_to_degrees(rad): return math.degrees(rad)

# Main Calculator Function
def calculator():
    operations = {
        '1': ("Addition", add, 2),
        '2': ("Subtraction", subtract, 2),
        '3': ("Multiplication", multiply, 2),
        '4': ("Division", divide, 2),
        '5': ("Square", square, 1),
        '6': ("Square Root", square_root, 1),
        '7': ("Degrees to Radians", degrees_to_radians, 1),
        '8': ("Radians to Degrees", radians_to_degrees, 1),
        '9': ("Exit", None, 0)
    }

    while True:
        for key, (desc, _, _) in operations.items():
            print(f"{key}. {desc}")

        choice = input("Enter your choice (1-9): ")
        if choice == '9':
            print("Exiting calculator. Goodbye!")
            break
        elif choice in operations:
            desc, func, num_args = operations[choice]

            try:
                nums = [float(input(f"Enter number {i+1}: ")) for i in range(num_args)]
                print("Result:", func(*nums))
            except ValueError:
                print("Invalid input! Please enter numeric values.")
        else:
            print("Invalid choice! Please enter a number between 1 and 9.")

# Run the calculator
calculator()
```

# Question
### Code a simple GUI-based calculator.

```python
import tkinter as tk

class SimpleCalculator:
    def __init__(self, root):
        self.root = root
        self.root.title("Calculator")

        # Entry widget to show the input and output
        self.entry = tk.Entry(self.root, width=16, font=("Arial", 24), borderwidth=2, relief="solid", justify="right")
        self.entry.grid(row=0, column=0, columnspan=4)

        # Buttons
            # Row 1
        self.create_button("7", 1, 0)
        self.create_button("8", 1, 1)
        self.create_button("9", 1, 2)
        self.create_button("/", 1, 3)
            # Row 2
        self.create_button("4", 2, 0)
        self.create_button("5", 2, 1)
        self.create_button("6", 2, 2)
        self.create_button("*", 2, 3)
            # Row 3
        self.create_button("1", 3, 0)
        self.create_button("2", 3, 1)
        self.create_button("3", 3, 2)
        self.create_button("-", 3, 3)
            # Row 4
        self.create_button("0", 4, 0)
        self.create_button(".", 4, 1)
        self.create_button("=", 4, 2)
        self.create_button("+", 4, 3)
            # Row 5
        self.create_button("C", 5, 0, 4) # Clear button (spans 2 columns)

    def create_button(self, text, row, column, colspan=1):
        button = tk.Button(self.root, text=text, width=5, height=2, font=("Arial", 18), command=lambda t=text: self.on_button_click(t))
        button.grid(row=row, column=column, columnspan=colspan, padx=5, pady=5)

    def on_button_click(self, button_text):
        if button_text == "=":
            try:
                expression = self.entry.get()
                result = eval(expression)
                self.entry.delete(0, tk.END)
                self.entry.insert(tk.END, str(result))
            except Exception as e:
                self.entry.delete(0, tk.END)
                self.entry.insert(tk.END, "Error")
        elif button_text == "C":
            self.entry.delete(0, tk.END)
        else:
            current_text = self.entry.get()
            self.entry.delete(0, tk.END)
            self.entry.insert(tk.END, current_text + button_text)

# Create the main window
root = tk.Tk()

# Create the calculator object and pass the window (root)
calculator = SimpleCalculator(root)

# Start the GUI event loop
root.mainloop()
```

# Question

### Code a program which can calculate the square roots of a user-defined quadratic equations.

```python
import cmath

# Function to calculate the roots of the quadratic equation
def find_roots(a, b, c):
    discriminant = b**2 - 4*a*c

    root1 = (-b + cmath.sqrt(discriminant)) / (2 * a)
    root2 = (-b - cmath.sqrt(discriminant)) / (2 * a)

    return root1, root2

# Equation formatting function
def formatEquation(a, b, c):
    terms = []

    # Handling coefficient of x^2
    if a == 1:
        terms.append("x^2")
    elif a == -1:
        terms.append("-x^2")
    else:
        terms.append(f"{int(a) if a.is_integer() else a}x^2")

    # Handling coefficient of x
    if b > 0:
        terms.append(f"+ {int(b) if b.is_integer() else b}x")
    elif b < 0:
        terms.append(f"- {abs(int(b) if b.is_integer() else b)}x")

    # Handling constant term
    if c > 0:
        terms.append(f"+ {int(c) if c.is_integer() else c}")
    elif c < 0:
        terms.append(f"- {abs(int(c) if c.is_integer() else c)}")

    return " ".join(terms)

# Main function
def main():
    a = float(input("Enter the coefficient of x^2: "))
    b = float(input("Enter the coefficient of x: "))
    c = float(input("Enter the coefficient of the constant term: "))

    if a == 0:
        print("This is not a quadratic equation (a cannot be zero).")
        return

    # Get the roots
    root1, root2 = find_roots(a, b, c)

    # Format output
    equation = formatEquation(a, b, c)

    # Display results
    print(f"Quadratic Equation: {equation} = 0")

    # Convert roots to int if possible
    root1 = int(root1.real) if root1.imag == 0 and root1.real.is_integer() else root1
    root2 = int(root2.real) if root2.imag == 0 and root2.real.is_integer() else root2

    print(f"The roots of the equation are: {root1} and {root2}")

# Run the program
main()
```

# Homework Question

### Code a program for solving simultaneous equations.

```python
import numpy as np

# Ask the user for the size of the matrix
n = int(input("Enter the number of equations (size of the square matrix A): "))

# Input matrix A
A = []
for i in range(n):
	row = list(map(float, input(f"Enter row {i+1} (space-separated): ").split()))
	A.append(row)
A = np.array(A)
print(f"Enter the coefficients matrix A (row by row):\n{A}")

# Input matrix B
B = []
for i in range(n):
	value = float(input(f"Enter value {i+1}: "))
	B.append(value)
B = np.array(B)
print(f"Enter the constants matrix B:\n{B}")

# Solve the equations
solution = np.linalg.solve(A, B)

# Display the solution
print("Solution:", solution)
```
